```python
candles = pd.read_csv('candles.csv')
```

# 5  20  
##  

```python
ma_s = ts.get_ma(
    candles['trade_price'],
    days = 5
)

ma_l = ts.get_ma(
    candles['trade_price'],
    days = 20
)

candles['ma_s'] = ma_s
candles['ma_l'] = ma_l

profit_range = candles.loc[
    candles['ma_s'] >= candles['ma_l']
]
```

```python
plt.figure(
    figsize = (12,8)
)

plt.scatter(
    profit_range.index,
    profit_range['trade_price'],
    color='y',
    s=15,
    marker='x'
)

plt.plot(
    candles['trade_price']
)

plt.plot(
    ma_l
)

# plt.plot(
#     candles['trade_price']
# )

plt.legend(
    ['hold point', 'trade_price', 'ma_long']
)
```
